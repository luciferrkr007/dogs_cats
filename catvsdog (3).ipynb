```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
import os
import random
```

```python
DIRECTORY=r'C:\Users\kumar\Downloads\archive\dogscats\train'
CATEGORIES=['cats','dogs']
```

```python
IMG_SIZE=100
data=[]

for category in CATEGORIES:
    folder=os.path.join(DIRECTORY,category)
    label=CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path=os.path.join(folder,img)
        img_arr=cv2.imread(img_path)
        img_arr=cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        data.append([img_arr,label])
        
```

```python
len(data)
```

```python
x=[]
y=[]
for feature,label in data:
    x.append(feature)
    y.append(label)
```

```python
X=np.array(x)
y=np.array(y)
```

```python
print(len(X),len(y))
```

```python
X=X/255
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,Dense,MaxPooling2D,Flatten
```

```python
model=Sequential()

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Flatten())

model.add(Dense(128,input_shape=X.shape[1:],activation='relu'))

model.add(Dense(128,activation='relu'))

model.add(Dense(2,activation='softmax'))

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

```

```python
model.fit(X,y,epochs=5,validation_split=0.1)
```

```python
def dog_cat_prediction(img):
    img=np.float32(img)
    pred_result=model.predict(img)
    return pred_result
    
    
```

```python
test_directory=r'Downloads\dog_cat_testing'
os.listdir(test_directory)
```

```python
X_test=[]
for img in os.listdir(test_directory):
    test_img=os.path.join(test_directory,img)
    test_img=cv2.imread(test_img)/255
    test_img=cv2.resize(test_img,(IMG_SIZE,IMG_SIZE))
    test_img.astype('float32')
    X_test.append(test_img)
len(X_test)
X_test=np.array(X_test)
```

```python
pred_result=dog_cat_prediction(X_test)
print(pred_result)
```

```python
i=1
rss=[]
for x,y in pred_result:
    if x>y:
        print(str(i) + ' cat')
        rss.append('cat')
    else:
        print(str(i) + ' dog')
        rss.append('dog')
    i+=1
```

```python
i=0
for img in X_test:
    plt.imshow(img)
    plt.title(rss[i])
    plt.show()
    i+=1
```
